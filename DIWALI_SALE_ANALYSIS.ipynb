```python
EXPLORATORY DATA ANALYSIS
```

```python
ON THE BASIS OF GENDER
```

```python
df.groupby(['Gender'])['Amount'].sum().reset_index()
```

```python
df.groupby(['Gender'])['Amount'].sum().plot(kind='bar',color='silver')
plt.xlabel('Gender')

plt.show()
```

**CONCLUSION= based on gender, females are spending 2.3X compare with the males**																		






                              **ON THE BASIS OF AGE**




```python
plt.hist(df['Age Group'].sort_values(ascending=True))
plt.show() 
```

```python
df.groupby(['Age Group'])['Amount'].sum().reset_index()
```

```python
df.groupby(['Age Group','Gender'])['Amount'].sum().sort_values(ascending=True).reset_index()
```

```python
df.groupby(['Age Group','Gender'])['Amount'].sum().sort_values(ascending=True).unstack().plot(kind='bar')
plt.xlabel('Age Group_gender')
plt.ylabel('Amount_sum')
plt.show()
```

**CONCLUSION= EXPENDITURE DONE BY POPULATION WITH AGE GROUP 26-35,FEMALE IS MAXIMUM,29% ON DIWALI AND 55+ AGE IS MINUMUM AROUND 4%**																				   

                                          **ON THE BASIS OF STATE**




```python
df.groupby(['State'])['Amount'].sum()
```

```python
df.groupby(['State'])['Amount'].sum().sort_values(ascending=True).plot(kind='bar',color='brown')
plt.xlabel('State')
plt.ylabel('Amount_sum')
plt.show()
```

```python
df.groupby(['State'])['Amount'].mean().reset_index()
```

```python
df.groupby(['State'])['Amount'].mean().sort_values(ascending=True).plot(kind='bar',color='brown')
plt.xlabel('State')
plt.ylabel('Amount_mean')
plt.show()
```

**CONCLUSION= EXPENDITURE DONE PER PERSON IN DELHI IS MAXIMUM, AROUND 7.27% ON DIWALI AND KARNATAKA IS ON SECOND NO. WITH 7.17%**
